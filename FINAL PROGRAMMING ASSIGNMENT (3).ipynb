# 3 Data Type Correction (Numeric vs Categorical)

```python
# Convert these from numeric codes into categorical
cat_cols = ["waterfront", "view", "condition", "grade"]
for c in cat_cols:
    df2[c] = df2[c].astype("Int64").astype("category")

# Keep these numeric
df2["bedrooms"]  = pd.to_numeric(df2["bedrooms"], errors="coerce")
df2["bathrooms"] = pd.to_numeric(df2["bathrooms"], errors="coerce")  # fractions matter
df2["floors"]    = pd.to_numeric(df2["floors"], errors="coerce")

df2.dtypes
```

# 5.Renovation Flag + Construction Decade

```python
df2["renovated"] = pd.to_numeric(df2["renovated"], errors="coerce").astype("Int64")
df2["renovated_flag"] = (df2["renovated"].fillna(0) != 0).astype(int).astype("category")

df2["built"] = pd.to_numeric(df2["built"], errors="coerce").astype("Int64")
df2["built_decade"] = (df2["built"] // 10 * 10).astype("Int64").astype("category")

df2[["built","built_decade","renovated","renovated_flag"]].head()
```

# 6 Missing Values Fill (Median for Numeric)

```python
missing_before = int(df2.isna().sum().sum())
print("Missing cells BEFORE:", missing_before)

num_cols = df2.select_dtypes(include=[np.number]).columns
cat_cols2 = df2.select_dtypes(include=["category"]).columns

# Numeric â†’ median
for c in num_cols:
    if df2[c].isna().any():
        df2[c] = df2[c].fillna(df2[c].median())

# Categorical â†’ Unknown
for c in cat_cols2:
    if df2[c].isna().any():
        df2[c] = df2[c].cat.add_categories(["Unknown"]).fillna("Unknown")

missing_after = int(df2.isna().sum().sum())
print("Missing cells AFTER:", missing_after)
```

# Part 3 - Outliers + Indicators

# 1. IQR Test (Keep rows, cap values in *_iqr)

```python
def iqr_cap(series, k=1.5):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lo = q1 - k*iqr
    hi = q3 + k*iqr
    capped = series.clip(lo, hi)
    return capped, lo, hi

iqr_summary = {}

for col in ["price", "bedrooms", "bathrooms", "sqft_living", "sqft_total"]:
    capped, lo, hi = iqr_cap(df2[col].astype(float))
    df2[col + "_iqr"] = capped
    outliers = int(((df2[col] < lo) | (df2[col] > hi)).sum())
    iqr_summary[col] = {"IQR_low": float(lo), "IQR_high": float(hi), "outlier_count": outliers}

pd.DataFrame(iqr_summary).T
```

# 15. Floor Space SQFT vs Price (GBP) Scatter â€” Before vs After + Regression Line (After)

```python
x = df2["sqft_living"].astype(float)
y = df2["price"].astype(float)

_, x_lo, x_hi = iqr_cap(x)
_, y_lo, y_hi = iqr_cap(y)

mask = (x >= x_lo) & (x <= x_hi) & (y >= y_lo) & (y <= y_hi)
x2, y2 = x[mask], y[mask]

coef = np.polyfit(x2, y2, 1)
xs = np.linspace(x2.min(), x2.max(), 200)
ys = coef[0]*xs + coef[1]

# bubble size = bedrooms
sizes = df2["bedrooms"].astype(float) * 20

plt.figure(figsize=(10,4))

# ---------- BEFORE ----------
plt.subplot(1,2,1)
sc1 = plt.scatter(x, y,
                  s=sizes,
                  c=df2["grade"].cat.codes,
                  alpha=0.5,
                  cmap="viridis")

plt.title("Floor Space vs Price (GBP) - Before")
plt.xlabel("sqft_living")
plt.ylabel("Price (GBP)")
plt.grid(True, alpha=0.3)

cbar = plt.colorbar(sc1)
cbar.set_label("House Grade")

# ---------- AFTER ----------
plt.subplot(1,2,2)
sc2 = plt.scatter(x2, y2,
                  s=sizes[mask],
                  c=df2.loc[mask, "grade"].cat.codes,
                  alpha=0.5,
                  cmap="plasma")

plt.plot(xs, ys, linewidth=2, label="Regression line")

plt.title("Floor Space vs Price (GBP) - After (Outliers Removed)")
plt.xlabel("sqft_living")
plt.ylabel("Price (GBP)")
plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.savefig("Floor space vs price.png",dpi=1000)
plt.show()

print("Regression: price â‰ˆ", round(coef[0],2), "* sqft_living +", round(coef[1],2))
print("Points kept:", int(mask.sum()), "out of", len(df2))
```
